### Detecting Outliers

```python
sns.boxplot(df['Age'])
plt.show()
sns.boxplot(df['Fare'])
```

```python
df=df[df['Age']<(df['Age'].mean()+3*df['Age'].std())]
df.shape
```

```python
sns.boxplot(df['Age'])
```

```python
q1=np.percentile(df['Fare'],25)
q3=np.percentile(df['Fare'],75)

outlier_low=q1-1.5*(q3-q1)
outlier_high=q1+1.5*(q3-q1)

df=df[(df['Fare']>outlier_low)&(df['Fare']<outlier_high)]
```

```python
sns.boxplot(df['Fare'])
```
