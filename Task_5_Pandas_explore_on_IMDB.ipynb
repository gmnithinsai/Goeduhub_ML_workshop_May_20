__8. Dealing with missing values__

```python
df.isnull().sum()
```

we could see some missing values in Revenue attribute and metascore attribute

```python
print(df['Metascore'].mean())
print(df['Revenue (Millions)'].median())
```

lets fill the values with the mean and median values of the respective attributes

```python
df['Metascore'].fillna(58.99,inplace=True)
```

```python
df['Revenue (Millions)'].fillna(47.99,inplace=True)
```

```python
df.isnull().sum()
```

__Now we can see that we have replaced the null values with mean and median values and there are no null values now__
