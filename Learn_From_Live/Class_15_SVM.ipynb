```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
```

```python
df.head()
```

```python
df['Target']=cancer.target
```

```python
df.head()
```

```python
df['cancer class']=df.Target.apply(lambda x:cancer.target)
```

```python
df.head()
```

```python
df['cancer class']=df.Target.apply(lambda x:cancer.target_names[x])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['Target','cancer class'],axis='columns')
y=df.Target
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=5)
```

```python
len(xtrain)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
pred=model.predict(xtest)
from sklearn.metrics import confusion_matrix
cm=np.array(confusion_matrix(ytest,pred))
print(cm)
```

```python
cm.ndim
```

```python
modelc=SVC(C=1)
modelc.fit(xtrain,ytrain)
modelc.score(xtest,ytest)
```

```python
modelc=SVC(C=2)
modelc.fit(xtrain,ytrain)
modelc.score(xtest,ytest)
```

## Kernel

```python
modelkernel=SVC(kernel='linear')
modelkernel.fit(xtrain,ytrain)
modelkernel.score(xtest,ytest)
```
