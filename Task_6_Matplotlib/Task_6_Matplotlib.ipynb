```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]

plt.plot(Days,Speed)
plt.title("Car speed measurement")
plt.xlabel('Days')
plt.ylabel('Car Speed')
plt.show()
```

__5. Plot Simple bar chart showing popularity of Programming Languages.__

- Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

- Popularity = [56, 39, 34, 34, 29]

- Security = [44 ,36 ,55, 50, 42]

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function

```python
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [56, 39, 34, 34, 29]
security = [44 ,36 ,55, 50, 42]
#plt.subplot(1,1,1)
plt.figure(figsize=(10,5))
plt.bar(languages,popularity,color='lightgreen',width=0.4)
plt.title('Languages and their popularity')
plt.show()
plt.figure(figsize=(10,5))
plt.bar(languages,security,color='lightblue',width=0.4)
plt.title('Languages and security')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.barh(languages,popularity,color='lightgreen')
plt.title('Languages and their popularity')
plt.show()
plt.figure(figsize=(10,5))
plt.barh(languages,security,color='lightblue')
plt.title('Languages and security')
plt.show()
```

__6.  Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.__

Marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]

Histogram showing Below Average, Average and Execellent distribution

```python
marks = np.array([ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])
plt.figure(figsize=(12,6))
plt.hist(marks,color='lightblue',edgecolor='blue')
plt.xlabel('bins')
plt.ylabel('Marks')
plt.title('Histogram of marks')
plt.vlines([40,60,80,100],ymin=0,ymax=4,linestyle='dashed')
```
